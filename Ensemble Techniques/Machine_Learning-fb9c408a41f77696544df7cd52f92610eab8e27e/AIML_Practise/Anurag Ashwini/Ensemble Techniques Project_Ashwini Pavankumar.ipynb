# Data Cleansing

## This includes missing value treatment, dropping of irrelevant attributes and converting the data into binary integers.

```python
# Checking for missing values
tel_data.isnull().sum()
```

## Missing values treatment
### The above output confirms that there are 11 missing records in the 'TotalCharges' column. Let us replace the null values with the median during data cleansing process.

## Label Encoding

###  The features gender, Partner, Dependents, PhoneService, PaperlessBilling, Churn have to be converted to binary integers because each of these take on two values( yes or no).

## One-Hot Encoding

###  The features like MultipleLines,InternetService, OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport, StreamingTV, StreamingMovies, Contract and PaymentMethod  and NumberOfAdditionalServices are not ordinal but take on more than two values. Hence one-hot encoding is used for these features.

```python
def data_cleansing(tel_data):
    
    # Impute missing values on TotalCharges
    
    tel_data.TotalCharges = tel_data.TotalCharges.replace(to_replace=np.nan, value= tel_data.TotalCharges.median())
    
    # Drop customer id
    tel_data.drop(['customerID'], axis=1, inplace=True)
   
    # Encode categorical features
    cat_cols = [c for c in tel_data.columns if tel_data[c].dtype == 'object' or c == 'SeniorCitizen']
    for col in cat_cols:
        if tel_data[col].nunique() == 2:
            tel_data[col], _ = pd.factorize(tel_data[col])
        else:
            tel_data = pd.get_dummies(tel_data, columns=[col])
    features = [c for c in tel_data.columns ]
    return tel_data[features]   

tel_data=data_cleansing(tel_data)
```

### We can see that some of the features  do not contribute to the churn prediction like 'customerID', 'OnlineSecurity_No internet service', 'OnlineBackup_No internet service','DeviceProtection_No internet service', 'TechSupport_No internet service','StreamingTV_No internet service', 'StreamingMovies_No internet service' and 'PhoneService'. Hence, they can be dropped from the data.

```python
def dropping_features(tel_data):
    drop_features = ['OnlineSecurity_No internet service', 'OnlineBackup_No internet service',
                     'DeviceProtection_No internet service', 'TechSupport_No internet service',
                     'StreamingTV_No internet service', 'StreamingMovies_No internet service',
                     'PhoneService']
    features = [c for c in tel_data.columns if c not in drop_features]
    return tel_data[features]
tel_data=dropping_features(tel_data)
```

# Data Preprocessing

### Segregating the data


```python
from sklearn.model_selection import train_test_split

X = tel_data.drop('Churn',axis=1)     # Predictor feature columns 
Y = tel_data['Churn']   # Predicted class 
```

### Spliting the data into Train and Test datasets ( 70:30 )

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

### Checking for the target balancing in the data

```python
# Number of customers and percentage of customers who churned in the original, train and test datasets 
churn = len(tel_data.loc[tel_data['Churn'] == 1])
no_churn = len(tel_data.loc[tel_data['Churn'] == 0])
print("Number of customers in the original data who have churned: {0} ({1:2.2f}%)".format(churn, (churn / (churn + no_churn)) * 100 ))
print("Number of customers in the original data who have not churned : {0} ({1:2.2f}%)".format(no_churn, (no_churn / (churn + no_churn)) * 100))
print("")
churn_train = len(y_train[y_train[:] == 1])
no_churn_train = len(y_train[y_train[:] == 0])
print("Number of customers in the training dataset who have churned: {0} ({1:2.2f}%)".format(churn_train, (churn_train / (churn_train + no_churn_train)) * 100 ))
print("Number of customers in the training dataset who have not churned : {0} ({1:2.2f}%)".format(no_churn_train, (no_churn_train / (churn_train + no_churn_train)) * 100))
print("")
churn_test = len(y_test[y_test[:] == 1])
no_churn_test = len(y_test[y_test[:] == 0])
print("Number of customers in the testing dataset who have churned : {0} ({1:2.2f}%)".format(churn_test, (churn_test / (churn_test + no_churn_test)) * 100 ))
print("Number of customers in the testing dataset who have not churned: {0} ({1:2.2f}%)".format(no_churn_test, (no_churn_test / (churn_test + no_churn_test)) * 100))
```

### The above output shows that the percentage of customers who have churned in the original data is 26.54%. We can observe target imbalance.

### We can observe from the above output that the percentages of customers who have churned in  training and test dataset are 27.20% and 24.99% respectively. Hence, we can say that the data is split well between the training and testing sets, but there is a target imbalance in the data which needs to be fixed.

### When the dataset does not represent all classes of data equally, the model might overfit to the class that’s represented more in the dataset and become oblivious to the existence of the minority class. It might even give a good accuracy but may fail miserably in real life. Hence we have to  fix the imbalance of the training data for better model fitting.
### We can use  a combination of a RandomUnderSampler and  SMOTE (over sampling) to balance the data.



```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline


rus=RandomUnderSampler(sampling_strategy=0.5)
sm = SMOTE(random_state=1,sampling_strategy=1.0)
steps=[('r',rus),('s',sm)]
pipeline=Pipeline(steps=steps)
x_train, y_train = pipeline.fit_resample(x_train, y_train)
x_train.shape, y_train.shape

```

```python
y_train = pd.DataFrame(y_train, columns = ['Churn'])
y_train.Churn.value_counts()
```

# Decision tree model

```python
# import necessary libraries
from sklearn import tree
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
dTree = DecisionTreeClassifier(criterion = 'entropy', random_state=1)
dTree.fit(x_train, y_train)
```

```python
dTree_train_acc=dTree.score(x_train, y_train)
dTree_acc=dTree.score(x_test, y_test)
print("The accuracy score of training data: {0} ".format(dTree_train_acc))
print("The accuracy score of test data: {0} ".format(dTree_acc))


```

### We can see that the accuracy on training data is almost 1, but the accuracy on test data is 0.72. Hence the decision tree needs to be pruned to avoid overfitting. Lets restrict the max depth to 3 for pruning.

## Reducing Overfitting(Regularization)

```python
# lets prune the decision tree to max_depth=3
dTreeR = DecisionTreeClassifier(criterion = 'entropy', max_depth = 9, random_state=1)
dTreeR.fit(x_train, y_train)
y_predict = dTreeR.predict(x_test)

dTreeR_train_acc=dTreeR.score(x_train, y_train)
dTreeR_acc=dTreeR.score(x_test, y_test)
print("The accuracy score of training data: {0} ".format(dTreeR_train_acc))
print("The accuracy score of test data: {0} ".format(dTreeR_acc))

# Confusion Matrix
cm=metrics.confusion_matrix(y_test, y_predict, labels=[1, 0])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True, fmt='g');
```

### We can see that the accuracy has improved after regularization.

## Confusion matrix
### True Positives (TP): We correctly predicted that 215 customers churned.
### True Negatives (TN): We correctly predicted that 1471 customers did not churn.
### False Positives (FP): We incorrectly predicted that 114 customers churned.("Type I error").
### False Negatives (FN): We incorrectly predicted 313 customers did not churn.( "Type II error").
