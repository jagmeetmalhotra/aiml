2 **Question: An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.**

A. Probability that none of the items are defective?

B. Probability that exactly one of the items is defective?

C. Probability that two or fewer of the items are defective?

D. Probability that three or more of the items are defective ?

**Solution**




````python
# A. Probability that none of the items are defective?
k=0
n=10
p=0.05
binomial=stats.binom.pmf(k,n,p)
binomial
````

````python
# B. Probability that exactly one of the items is defective?
k=1
n=10
p=0.05
binomial=stats.binom.pmf(k,n,p)
binomial
````

````python
# C. Probability that two or fewer of the items are defective?
k=[0,1,2]
n=10
p=0.05
binomial=stats.binom.pmf(k,n,p)
sum(binomial)
````

````python
# D. Probability that three or more of the items are defective ?

#this is the inverse of the above question. Therefore probability will be 1-the above answer
p=1-sum(binomial)
p
````

5 **Question: A group of 300 professionals sat for a competitive exam. The results show the information of marks obtained by them have a mean of 60 and a standard deviation of 12. The pattern of marks follows a normal distribution. Answer the following questions.**

A. What is the percentage of students who score more than 80.

B. What is the percentage of students who score less than 50.

C. What should be the distinction mark if the highest 10% of students are to be awarded distinction?

**Solution**




````python
#A. What is the percentage of students who score more than 80.
z=(80-60)/(np.sqrt(2)*12)
result=stats.norm(0,1).cdf(z)
(1-result)*100
````

````python
#B. What is the percentage of students who score less than 50.

z=(50-60)/(np.sqrt(2)*12)
result=stats.norm(0,1).cdf(z)
result*100
````

````python
# C. What should be the distinction mark if the highest 10% of students are to be awarded distinction?

#percentage cutoff for top 10% of students also means finding highest score of lowest 90% students

cutoff_norm=stats.norm.ppf(0.9)
cutoff=cutoff_norm*np.sqrt(2)*12+60
cutoff
````

````python
# Drop the null values
data.dropna(inplace=True)
data.reset_index(inplace=True, drop=True)
data.info()
````

````python
# Convert the ‘Funding’ features to a numerical value.

# checking if all funding values start with a $ sign
# also to check the different last charachters such as B, K, M
x=set()
y=set()
for i in data.Funding:
    x.add(i[-1])
    y.add(i[0])
    
print(x)
print(y)
````

````python
#Removing the $ sign and converting all values in millions

fund_col=data.Funding.values

for i in range(len(fund_col)):
    fund_col[i]=fund_col[i][1:]
    
for i in range(len(fund_col)):
    
    if fund_col[i][-1]=='B':          #checking if the last charachter is B, K or M
        fund_col[i]=float(fund_col[i][:-1])*(10**3)              # converting the number into millions
    
    elif fund_col[i][-1]=='K':
        fund_col[i]=float(fund_col[i][:-1])/(10**3)               # converting the number into millions 
    
    elif fund_col[i][-1]=='M':
        fund_col[i]=float(fund_col[i][:-1])                     # converting the number into millions
````

````python
#Get the lower fence from the box plot.
q1=data.Funding.quantile(0.25)
q3=data.Funding.quantile(0.75)
dmax=data.Funding.max()
dmin=data.Funding.min()
iqr=q3-q1
upper_fence=min(q3+(1.5*iqr),dmax)
lower_fence=max(q1-(1.5*iqr),dmin)

lower_fence
````

````python
# Check number of outliers greater than upper fence
np.sum(data.Funding.values>upper_fence)
````

````python
# • Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?
# Write the null hypothesis and alternative hypothesis.
# Test for significance and conclusion
````

````python
# H0(null hypothesis): there is no significant difference between Funds raised by companies that are still operating vs companies that closed down
# H1(alternative hypothesis): there is a difference
#significance value = 0.05
from scipy.stats import ttest_ind
x=data['Funding'].loc[data['OperatingState']=='Operating']
y=data['Funding'].loc[data['OperatingState']=='Closed']

t_statistic, p_value = ttest_ind(x, y)
print(t_statistic, p_value)
````

````python
#since p value is less than singnificance level
# conclusion is that we have failed to reject the null hypothesis. 
#(there is no significant difference between Funds raised by companies that are 
#still operating vs companies that closed down)
````

````python
# Write your hypothesis comparing the proportion of companies that are operating between winners and contestants:
#  Write the null hypothesis and alternative hypothesis.
#  Test for significance and conclusion
````

````python
#H0 null hypothesis: there is no difference in proportion of companies that are operating between winners and contestants
#H1 alternative hypothesis: there is a significant difference in proportion of companies that are operating between winners and contestants

#significance level = 0.05

winners_operating=df[df['Result']=='Winner'].OperatingState.value_counts()[0]            #number of winners who are operating
contestants_operating=df[df['Result']=='Contestant'].OperatingState.value_counts()[0]    #number of contestants who are operating
n_winners=df.Result.value_counts()[3]                                                    #number of winners
n_contestants=df.Result.value_counts()[0]                                                #number of contestants

from statsmodels.stats.proportion import proportions_ztest

stat, pval = proportions_ztest([winners_operating, contestants_operating] , [n_winners, n_contestants])

if pval < 0.05:
    print(f'With a p-value of {round(pval,4)} the difference is significant. aka |We reject the null|')
else:
    print(f'With a p-value of {round(pval,4)} the difference is not significant. aka |We fail to reject the null|')
````

````python
#H0(null hypothesis): there is no significant difference between funds raised by companies across NY, SF and EU events from 2013 onwards.
# H1(alternative hypothesis): there is a significant difference between funds raised by companies across NY, SF and EU events from 2013 onwards.
# significance level = 0.05

import scipy.stats as stats

stats.f_oneway(df_city_NY.Funding, df_city_SF.Funding, df_city_EU.Funding)
````
