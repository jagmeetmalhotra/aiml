#### 3. Clustering: [15 Marks]
##### A. Apply K-Means clustering for 2 to 10 clusters. [3 Marks]
##### B. Plot a visual and find elbow point. [2 Marks]
##### C. On the above visual, highlight which are the possible Elbow points. [1 Marks]
##### D. Train a K-means clustering model once again on the optimal number of clusters. [3 Marks]
##### E. Add a new feature in the DataFrame which will have labels based upon cluster value. [2 Marks]
##### F. Plot a visual and color the datapoints based upon clusters. [2 Marks]
##### G. Pass a new DataPoint and predict which cluster it belongs to. [2 Marks]

```python
# PartA.3.A. Apply K-Means clustering for 2 to 10 clusters. [3 Marks]

from sklearn.cluster import KMeans
df=analysis_df.copy()
#Dropping non necessary columns
df.drop('yr', axis='columns', inplace=True)
df.drop('origin', axis='columns', inplace=True)
df.drop('car_name', axis='columns', inplace=True)


features = list(["cyl", "disp", "hp", "wt", "acc"])
y = df["mpg"]
X = df[features]

# WCSS is the sum of squared distance between each point and the centroid in a cluster
wcss = []
for i in range(2,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# PartA.3.B. Plot a visual and find elbow point. [2 Marks]
plt.plot(range(2,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### PartA.3.C. On the above visual, highlight which are the possible Elbow points. [1 Marks]
##### Ans  :The possible elbow point are 3 and 4.
##### Should model based on that. We will continue with 3 as the elbow point.
##### Else we may lead in a overfit zone

```python
#PartA.3.D. Train a K-means clustering model once again on the optimal number of clusters. [3 Marks]
kmeans = KMeans(n_clusters=3,init='k-means++',random_state=0) 
y_kmeans = kmeans.fit_predict(X)
```

```python
#PartA.3.E. Add a new feature in the DataFrame which will have labels based upon cluster value. [2 Marks]
df["Cluster"]= y_kmeans
```

```python
df.head()
```

```python
#PartA.3.F. Plot a visual and color the datapoints based upon clusters.
sns.pairplot(df,hue="Cluster", palette= ["red", "green", "blue"])
```

```python
#PartA.3.G. Pass a new DataPoint and predict which cluster it belongs to. [2 Marks]
# Since we took 5 feature model 
#"cyl" = 4 
#"disp" = 100 
#"hp" = 90 
#"wt" = 2500
#"acc" = 13

new_car = np.array([4, 100, 90, 2500, 13])
prediction = kmeans.predict([new_car])
print(prediction)
```

```python
#As per the prodiction the car belongs to group 1. which is correct
```

```python
# Importing libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from scipy.stats import zscore
from sklearn.decomposition import PCA
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
```

```python
# PartB.1.A Read ‘vehicle.csv’ and save as DataFrame. [1 Marks]
vehicle_df = pd.read_csv('vehicle.csv')
vehicle_df.head()
```

```python
# PartB.1.B. Check percentage of missing values and impute with correct approach. [1 Marks]
vehicle_df.info()
```

```python
#Cheking null values 
vehicle_df.isnull().sum()
```

```python
# replacing it with respectiv medians
vehicle_df= vehicle_df.fillna(vehicle_df.median())
```

```python
#verifying no null values are left 
vehicle_df.isnull().sum()
```

```python
# PartB.1.C. Visualize a Pie-chart and print percentage of values for variable ‘class’.
pd.value_counts(vehicle_df["class"]).plot(kind="pie", autopct='%1.1f%%')
```

```python
# PartB.1.D. Check for duplicate rows in the data and impute with correct approach. [1 Marks]
duplicate = vehicle_df[vehicle_df.duplicated()]
  
print("Duplicate Rows :")
  
# Print the resultant Dataframe
duplicate
```

```python
# No duplicate rows found
```

```python
# Encoding Class car, bus and van to categorical type
```

```python
veh_dic = {"car":1,"bus":2,"van":3}
vehicle_df["class"].replace(to_replace = veh_dic,inplace=True)
```

```python
vehicle_df["class"].value_counts()
```

```python
# Data Preparation: [2 Marks]
# PartB.2.A. Split data into X and Y. [Train and Test optional] [1 Marks]
# PartB.2.B. Standardize the Data. [1 Marks]
# Solving both at same time, before splitting will standardize the X variables. 

X = vehicle_df.drop('class', axis = 1)
y = vehicle_df["class"]
X_std = X.apply(zscore)
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.20, random_state=101)
```

```python
#PArtB.3.A Train a base Classification model using SVM
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train,y_train)
```

```python
#PartB.3.B. Print Classification metrics for train data. [1 Marks]
from sklearn.metrics import accuracy_score, classification_report 
print("Train Score")
print(svc_model.score(X_train, y_train))
print("Test Score")
print(svc_model.score(X_test, y_test))

y_predict = svc_model.predict(X_test)

print("\nConfusion Matrix :")
label = ["car","bus","van"]
ax= plt.subplot()
sns.heatmap(metrics.confusion_matrix(y_test,y_predict),annot=True, ax = ax)
ax.set_xlabel('Predicted');
ax.set_ylabel('Actual'); 
ax.xaxis.set_ticklabels(label); 
ax.yaxis.set_ticklabels(label);
plt.show()
print("Classification Report :")
print(metrics.classification_report(y_test,y_predict,target_names = label))
```

```python
#PartB.3.C. Apply PCA on the data with 10 components.  
pca = PCA(n_components = 10)
pca.fit(X_std)
```

```python
#PartB.3.D. Visualize Cumulative Variance Explained with Number of Components. [2 Marks]
#PartB.3.E. Draw a horizontal line on the above plot to highlight the threshold of 90%. [1 Marks]
#Both are out in one graph.
plt.figure(figsize = (12,5))
plt.xticks(np.arange(0,11))
plt.step(list(range(1,11)),np.cumsum(pca.explained_variance_ratio_), where = 'mid')
plt.axhline(y = 0.9, color = 'r', linestyle = '-')
plt.ylabel('Cum of variation explained')
plt.xlabel('Eigen-Value')
plt.show()
```

```python
#PartB.3.F. Apply PCA on the data. This time Select Minimum Components with 90% or above variance explained.
# Form the red line we can see number of component for 90% variance explained in 5 and 6
# we will start with 5 and later also try for 5 
pca5 = PCA(n_components = 5)
pca5.fit(X_std)
print(pca5.explained_variance_ratio_)
```

```python
#PartB.3.G. Train SVM model on components selected from above step. [1 Marks]
# Using only those 5 important features to develop the model
X_pca5_transformed = pca5.transform(X_std)
X_train, X_test, y_train, y_test = train_test_split(X_pca5_transformed, y, test_size=0.2, random_state = 101)
svc_model_pca = SVC()
svc_model_pca.fit(X_train, y_train)
y_pred_pca = svc_model_pca.predict(X_test)
print("The accuracy score for the component is %.2f%%"%(svc_model_pca.score(X_test, y_test) * 100))
```

```python
#PartB.3.H. Print Classification metrics for train data of above model and share insights.


print("\nConfusion Matrix :")
label = ["car","bus","van"]
ax= plt.subplot()
sns.heatmap(metrics.confusion_matrix(y_test,y_pred_pca),annot=True, ax = ax)
ax.set_xlabel('Predicted');
ax.set_ylabel('Actual'); 
ax.xaxis.set_ticklabels(label); 
ax.yaxis.set_ticklabels(label);
plt.show()
print("Classification Report :")
print(metrics.classification_report(y_test,y_pred_pca,target_names = label))
```

##### -Variables included are metric level or nominal level.
##### -Features are low dimensional in nature.
##### -Independent variables are numeric in nature.


### Part B.5.B. Explain advantages and limitations of PCA. [3 Marks]
#### Advantages of PCA
##### -Removes multi-collinearity to improve ML model performance
##### -Help to reduce overfitting.
##### -Improves the data visualization.
##### -It also help to improve the performance of Algorithm
##### -Decreases storage requirements
##### -Avoids curse of dimensionality

#### Limitations of PCA
##### - Data normalization must be needed before applying PCA
##### - Some level of information loss.
##### - Independent variables are becoming less interpretable.
