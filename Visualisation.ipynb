```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt #seaborn is based on matplotlib
sns.set(color_codes=True) #add nice backgrounds to the graphs
%matplotlib inline 
#tells python to actually display the graphs
```

```python
auto=pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

# plot univariate distribution - one dimentional dataset
    The most convenient way to take a quick look at a univariate distribution in seaborn is the distplot(), By default this will draw a histogram and fit a kernel density edtimate KDE

```python
sns.distplot(auto['highway_mpg'])
```

```python
sns.distplot(auto['city_mpg'], kde=False, rug=True)
```

# Plotting bivariate distribution - two dimentional dataset
It can be useful to vihualise relationship between two variable. The easiest way to do this is to use joinplot() function, which creates a scatterplot of two variable with the histogram of each next to appropriate axes.

```python
sns.jointplot(auto['engine_size'], auto['horsepower']);
```

# Hex Bin plot
We can make a hex bin plot that breaks the 2D area into hexagon and number of points in each hexagon determines the color

```python
sns.jointplot(auto['engine_size'],auto['horsepower'],kind="hex");
```

# Kernel Density Estimation 
We can make a 2D estimation of the density also

```python
sns.jointplot(auto['engine_size'],auto['horsepower'],kind="kde");
```

# Visualizing pairwise relationships in a dataset
To plot multiple pair wise scatterplots in a distribution, you can use pairplot() function. This creates a matrices of axes and shows the realtionship of each pair of the columns in a dataframe. It also draws the histogram of each variable on the diagonal axes

```python
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']]);
```

#  Plotting categorical variable 
in a strip plot, the scatterplot will usually overlap. This makesit difficult to see the full distribution of data. One easy solution is to adjust the positions( only along the categorial axis) using "jitter"

```python
sns.stripplot(auto['fuel_type'], auto['horsepower'],jitter=True)
```

```python
sns.stripplot(auto['fuel_type'], auto['horsepower'])
```

```python
# another option is to use a swarmplot() which position each scatterplot point on the categorial 
# axis and avoids operlapping points
sns.swarmplot(auto['fuel_type'], auto['horsepower'])
```

# Boxplot
Another common graph is a boxplot(). This kind of plot shows three quartile values of the distribution along with extreme values. The "whiskers" extend to points that lie within 1.5 IQR of the lower and upper quartile, and then observations that fall outside this range are displayed independently

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'])
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type'])
```

# Barplot
We can plot the mean of the dataset, seperated by categories using the barplot() function. when there are multiple observations in each category, it uses  bootstrapping to compute a confidence interval around the estimate and plots that using error bars:
Bar plots start at 0, which may or may not be practical based on data

```python
sns.barplot(auto['body_style'], auto['horsepower'], hue=auto['fuel_type'])
```

# Histogram 
countplot

```python
sns.countplot(auto['body_style'],hue=auto['fuel_type'])
```

```python
sns.countplot(auto['body_style'])
```

# Point Plot
An alternative style of visualisizing

```python
sns.pointplot(auto['body_style'], auto['horsepower'], hue=auto['number_of_doors'])
```

# Mutipanel Categorial Plots


```python
sns.catplot(x="fuel_type",
           y="horsepower",
           hue="number_of_doors",
           col="drive_wheels",
           data=auto,
           kind="violin")
```

```python
sns.catplot(x="fuel_type",
           y="horsepower",
           hue="number_of_doors",
           col="drive_wheels",
           data=auto,
           kind="box")
```

# Linear Regression plot

```python
sns.lmplot(y="horsepower",x="engine_size",data=auto)
```

```python
sns.lmplot(y="horsepower",x="engine_size", hue="fuel_type", data=auto);
```
