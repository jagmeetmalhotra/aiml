```
def get_number():
    
    for n in range(100):
        
        return n
```

```
num_a = get_number()
print(num_a)
```

```
def generate_number():
    
    for n in range(100):
        
        yield n
```

```
num_b = generate_number()
print(num_b)
```

```
next(num_b)
```

```
def generate_number2():
    
    for n in range(100):
        
        if n % 2 == 0 :
            yield n
```

```
num_c = generate_number2()
```

```
next(num_c)
```
