### 4. Model Building: 6
a. Split data into X and Y. [1 Marks]

b. Split data into train and test with 80:20 proportion. [1 Marks]

c. Train a Supervised Learning Classification base model using KNN classifier. [2 Marks]

```python
from sklearn import preprocessing
from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Part1.4.a. Split data into X and Y. [1 Marks]
X=df_concat.drop(columns="Class")
Y=df_concat["Class"] # value to be calssified

# Standardisation / Scaling of Data
X_Scaled = X.apply(zscore)
X_Scaled.describe().T
```

```python
# Part1.4.b. Split data into train and test with 80:20 proportion. 
X_train, X_test, y_train, y_test = train_test_split(X_Scaled, Y, test_size=0.20, random_state=10,shuffle=True)

#Values in each set 
print("X Train ",X_train.shape[0])
print("--- X Test",X_test.shape[0])
print("y Train",y_train.shape[0])
print("--- y Test",y_test.shape[0])
```

```python
# Part1.4.c. Train a Supervised Learning Classification base model using KNN classifier. [2 Marks]
# Starting with 3 nearest neighbours
# loading library
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
# Learning model (k = 3)
knn = KNeighborsClassifier(n_neighbors = 3)
#Fitting the model
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
# Part1.4.d. Print all the possible classification metrics for both train and test data. [2 Marks]
print('Accuracy [Training Data]=:',knn.score(X_train, y_train) )
print('Accuracy [Test data ]   =:',knn.score(X_test, y_test) )
#Classification Report
print("===========================================")
print("KNN : Classification Report On Test data")
print("===========================================")
print(classification_report(y_test,y_pred))
```

```python
#Creating confusion matrix 
print("===========================================")
print("KNN : Confusion Matrix On Test data")
print("===========================================")
cm = confusion_matrix(y_test,y_pred)
df_cm = pd.DataFrame(cm, index = [i for i in ["Normal","Type_H","Type_S"]],
                  columns = [i for i in ["Pred Normal","Pred Type_H","Pred Type_S"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
plt.show()
```

## Observations 


#### Type_S is precision 96% which is good.
#### Recall is 82% for Type_S i.e. out of all Type_S model was able to identify 86% case 
#### Overall Accuracy of Model in 66% which is very low. Tuning will be required
#### Type H Precision and recall is very low i.e. 35% and 50% respectively
#### Normal Precision and recall is very low i.e. 47% for both

### 5. Performance Improvement: 4
a. Tune the parameters/hyperparameters to improve the performance of the base model. [2 Marks]

b. Clearly showcase improvement in performance achieved. [1 Marks]
For Example:

i. Accuracy: +15% improvement

ii. Precision: +10% improvement.

c. Clearly state which parameters contributed most to improve model performance.
What could be the probable reason? [1 Marks]

```python
#Part1.5.a. Tune the parameters/hyperparameters to improve the performance of the base model. [2 Marks]

# empty list that will hold accuracy scores
ac_scores = []
# creating odd list of K for KNN
myList = list(range(1,100))
# subsetting just the odd ones
neighbors = list(filter(lambda x: x % 2 != 0, myList))

# perform accuracy metrics for values from 1,3,5....
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    # predict the response
    y_pred = knn.predict(X_test)
    # evaluate accuracy
    scores = accuracy_score(y_test, y_pred)
    ac_scores.append(scores)

# changing to misclassification error
MSE = [1 - x for x in ac_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print("The optimal number of neighbors is %d" % optimal_k)
```

```python
import matplotlib.pyplot as plt
# Misclassification error vs K
plt.plot(neighbors, MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```

```python
# Part1.5.b. Clearly showcase improvement in performance achieved. [1 Marks] For Example:
# i. Accuracy: +15% improvement
# ii. Precision: +10% improvement.
    
#Use K=57 
knn = KNeighborsClassifier(n_neighbors = 57)

# fitting the model
knn.fit(X_train, y_train)

# predict the response
y_pred = knn.predict(X_test)

print('Accuracy [Training Data]=:',knn.score(X_train, y_train) )
print('Accuracy [Test data ]   =:',knn.score(X_test, y_test) )


#Classification Report
print("===========================================")
print("KNN Tuned Classification Report:Test data")
print("===========================================")
print(classification_report(y_test,y_pred))

```

```python
#Creating confusion matrix 
print("===========================================")
print("KNN Tuned Confusion Matrix:Test data")
print("===========================================")
cm = confusion_matrix(y_test,y_pred)
df_cm = pd.DataFrame(cm, index = [i for i in ["Normal","Type_H","Type_S"]],
                  columns = [i for i in ["Pred Normal","Pred Type_H","Pred Type_S"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
plt.show()
```

## Observations 
                 | Untuned(K=3)      |     Tuned(K=57)     | Description 

        Accuracy |    66%            |        81%          | 15% improvment in overall accuracy

        Precision|    71%            |        85%          | 14% uptick

     Type H Prec.|    35%            |        78%          | 43% uptick

     Normal Prec.|    47%            |        60%          | 13% uptick

        Recall   |    66%            |        81%          | 15% uptick

    Type H Recall|    50%            |        58%          | 8% uptick

    Normal Recall|    47%            |        88%          | 41% uptick 
    


With the inrease in number of Neighbours form 3 to 57 the model improved sigficicanlty especailly for classes previously classified incorrectly Normal and Type H. ## Part2.3. Data Preparation and model building:

```python
#Part2.3.a. Split data into X and Y. [1 Mark]
#[Recommended to drop ID & ZipCode. LoanOnCard is target Variable]

df_merged.drop('ID',axis=1,inplace=True)
df_merged.drop('ZipCode',axis=1,inplace=True)

X=df_merged.drop(columns="LoanOnCard")
Y=df_merged["LoanOnCard"] # value to be predicted
```

```python
#Part2.3.b. Split data into train and test. Keep 25% data reserved for testing. [1 Mark]
#Standardization of data
from sklearn import preprocessing 
from sklearn.preprocessing import StandardScaler 
X_std=pd.DataFrame(StandardScaler().fit_transform(X))
X_std.columns=X.columns

#Splitting data into train dataset and test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_std,Y,test_size=0.25,random_state=42,shuffle=True)
```

```python
#Part2.3.c. Train a Supervised Learning Classification base model - Logistic Regression. [2 Mark]
from sklearn import metrics
from sklearn.linear_model import LogisticRegression

# Fit the model on train
model = LogisticRegression()
model.fit(X_train, y_train)
#Predict on test
y_pred = model.predict(X_test)
```

```python
#Part2.3.d. Print evaluation metrics for the model and clearly share insights. [1 Mark]

# Logistic Regression Classification Reportand confusion Matrix
print("=========================================================")
print("Logistic Regression: Classification Report On Test data")
print("=========================================================")
print(classification_report(y_test,y_pred))

print("=========================================================")
print("Logistic Regression: Confusion Matrix")
print("=========================================================")

cm=metrics.confusion_matrix(y_test, y_pred, labels=[1, 0])
df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Loan_P 1","Loan_P 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True,fmt='g')
```

### Observation
#### Accuracy of the model is 95%
#### Recall of model is 0.66% for Loan = 1 caregory which is very low it means out of all the loans given we were able to predict only 66% of the cases which is very low. 
### Imbalanced data is reason for the same since in the sample number of Loans=1 category was very less.

```python
#Part2.3.e Balance the data using the right balancing technique. [2 Mark]
# Importing the RandomOverSampler
import imblearn
from imblearn.metrics import classification_report_imbalanced
from imblearn.over_sampling import RandomOverSampler

random_oversampler = RandomOverSampler(random_state=1)

# Fit the original training data to the random_oversampler model
X_rebalanced, y_rebalanced = random_oversampler.fit_resample(X_train, y_train)
```

```python
# Value count with LoanOnCard y rebalanced
y_rebalanced.value_counts()
```

```python
#Part2.3.f Again train the same previous model on balanced data.
model = LogisticRegression()
logistic_regression_rebalanced = model.fit(X_rebalanced, y_rebalanced)
# Re-Predict on test 
y_rebalanced_prediction = logistic_regression_rebalanced.predict(X_test)
```

```python
#Part2.3.g Print evaluation metrics and clearly share differences observed.
print("=========================================================")
print("Logistic Regression data balancing Classification Report ")
print("=========================================================")
print(classification_report(y_test,y_rebalanced_prediction))

print("=========================================================")
print("Logistic Regression data balancing Confusion Matrix")
print("=========================================================")

cm=metrics.confusion_matrix(y_test, y_rebalanced_prediction, labels=[1, 0])
df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Loan_P 1","Loan_P 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True,fmt='g')
```

### Observation after rebalncing

```python
# Part2.4.b. Tune parameters/hyperparameters for each of the models wherever required and finalize a model
# SVM Tuning 
from matplotlib.colors import Normalize
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV

class MidpointNormalize(Normalize):
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))


C_range = np.logspace(-2, 2, 5)
gamma_range = np.logspace(-2, 2, 5)
param_grid = dict(gamma=gamma_range, C=C_range)
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
grid = GridSearchCV(SVC(), param_grid=param_grid, cv=cv)
grid.fit(X_train, y_train)

print(
    "The best parameters are %s with a score of %0.2f"
    % (grid.best_params_, grid.best_score_)
)

# HeatMap
scores = grid.cv_results_["mean_test_score"].reshape(len(C_range), len(gamma_range))
plt.figure(figsize=(8, 6))
plt.subplots_adjust(left=0.2, right=0.95, bottom=0.15, top=0.95)
plt.imshow(
    scores,
    interpolation="nearest",
    cmap=plt.cm.hot,
    norm=MidpointNormalize(vmin=0.2, midpoint=0.92),
)
plt.xlabel("gamma")
plt.ylabel("C")
plt.colorbar()
plt.xticks(np.arange(len(gamma_range)), gamma_range, rotation=45)
plt.yticks(np.arange(len(C_range)), C_range)
plt.title("Validation accuracy")
plt.show()
```

```python
# Tuning for K nearest neighbours 
ac_scores = []
# creating odd list of K for KNN
myList = list(range(1,100))
# subsetting just the odd ones
neighbors = list(filter(lambda x: x % 2 != 0, myList))

# perform accuracy metrics for values from 1,3,5....
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    # predict the response
    y_pred = knn.predict(X_test)
    # evaluate accuracy
    scores = accuracy_score(y_test, y_pred)
    ac_scores.append(scores)

# changing to misclassification error
MSE = [1 - x for x in ac_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print("The optimal number of neighbors is %d" % optimal_k)
```

```python
import matplotlib.pyplot as plt
# Misclassification error vs K
plt.plot(neighbors, MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```
