# DOMAIN: Semiconductor manufacturing process

### • CONTEXT: A complex modern semiconductor manufacturing process is normally under constant surveillance via the monitoring of signals/variables collected from sensors and or process measurement points. However, not all of these signals are equally valuable in a specific monitoring system. The measured signals contain a combination of useful information, irrelevant information as well as noise. Engineers typically have a much larger number of signals than are actually required. If we consider each type of signal as a feature, then feature selection may be applied to identify the most relevant signals. The Process Engineers may then use these signals to determine key factors contributing to yield excursions downstream in the process. This will enable an increase in process throughput, decreased time to learning and reduce the per unit production costs. These signals can be used as features to predict the yield type. And by analysing and trying out different combinations of features, essential signals that are impacting the yield type can be identified.

### DATA DESCRIPTION: sensor-data.csv : (1567, 592) The data consists of 1567 datapoints each with 591 features. The dataset presented in this case represents a selection of such features where each example represents a single production entity with associated measured features and the labels represent a simple pass/fail yield for in house line testing. Target column “ –1” corresponds to a pass and “1” corresponds to a fail and the data time stamp is for that specific test point.

### PROJECT OBJECTIVE: We will build a classifier to predict the Pass/Fail yield of a particular process entity and analyse whether all the features are required to build the model or not

```python
# 2.A Write a for loop which will remove all the features with 20%+ Null values and impute rest with mean of the feature.
```

```python
signal_df.shape
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
thresh_perc = 0.20
columns_to_drop=[]
for i in range(1, (signal_df.shape[1])):
    colname = signal_df.columns[i]
    null_pct_in_col= signal_df.isnull().sum()[colname]/len(signal_df)
    if(null_pct_in_col >= thresh_perc ):
        columns_to_drop.append(colname)
    else:
        signal_df[colname] = imputer.fit_transform(signal_df[colname].values.reshape(-1,1))
        
        
print(columns_to_drop)
```

```python
signal_df = signal_df.drop(columns_to_drop, axis=1) 
```

```python
signal_df.shape
```

```python
# 32 columns have been dropped and for others mean have been implmented as the impute strategy 
```

```python
# 2.B. Identify and drop the features which are having same value for all the rows. [3 Marks]
# Idnetifying Columns with non unique values in all the rows
signal_df.columns[signal_df.nunique()==1]
```

```python
#dropping those columns i.e. only retaining those columns with non unique values > 1
signal_df=signal_df[signal_df.columns[signal_df.nunique()>1]]
```

```python
signal_df.shape
```

```python
#2.C Drop other features if required using relevant functional knowledge. Clearly justify the same. [2 Marks]
signal_df=signal_df.drop('Time',axis=1)
# Droped Time Columns as its not required to make any prediction
```

```python
signal_df.shape
```

```python
#2.D Check for multi-collinearity in the data and take necessary action. [3 Marks]
#checking for correlation 
#Assuming high colleniarity if corr > .7

# Create correlation matrix
signal_df_corr_mat = signal_df.corr().abs()

# Select top half of orrelation matrix
top = signal_df_corr_mat.where(np.triu(np.ones(signal_df_corr_mat.shape), k=1).astype(bool))

# Find features with correlation greater than 0.70 high multicollenarity 
features_to_drop = [column for column in top.columns if any(top[column] > 0.70)]

print(features_to_drop)
```

```python
# Drop high milticollenarity features 
signal_df=signal_df.drop(features_to_drop, axis=1)
```

```python
signal_df.shape
```

```python
# 2.E Make all relevant modifications on the data using both functional/logical reasoning/assumptions. 
#label encoding the Pass/Fail to 1 and 0 instead of -1 and 1 making it easy to understand 
signal_df['Pass/Fail']=signal_df['Pass/Fail'].replace([-1,1],[1,0])
```

```python
signal_df.describe().T
```

```python
#Handling Outliers 
# 202 columns is Pass Fail which is the variable to be preicted hence not including it in histograms.
#will look at 201th col seperately
plt.figure(figsize=(50, 50))
col = 1

for i in signal_df.columns[range(200)]:
    plt.subplot(20,10, col)
    sns.boxplot(data=signal_df[i],color='blue',orient='h')
    col += 1
```

```python
sns.boxplot(data=signal_df[signal_df.columns[200]],color='blue',orient='h')
```

```python
# Repalcing Outlisers with median value.
for i in signal_df.columns[range(200)]:
    q1 = signal_df[i].quantile(0.25)
    q3 = signal_df[i].quantile(0.75)
    iqr = q3 - q1
    
    low = q1 - 1.5 * iqr
    high = q3 + 1.5 * iqr
    
    signal_df.loc[(signal_df[i] < low) | (signal_df[i] > high), i] = signal_df[i].median()
```

```python
#4.A Segregate predictors vs target attributes. [2 Marks]
X = signal_df.drop('Pass/Fail',axis=1)
# X will be the feature used to predict

y = signal_df['Pass/Fail']
# y will be the target
```

```python
#Only 6.6% of class Fail(0), balancing the data using SMOTE
```

```python
from imblearn import over_sampling,under_sampling
from imblearn.over_sampling import SMOTE

#SMOTE oversampling 
SM = over_sampling.SMOTE()
#Updated data
X_SM,y_SM = SM.fit_resample(X,y)
```

```python
y_SM.value_counts()
```

```python
# 4.C. Perform train-test split and standardise the data or vice versa if required. [3 Marks]
```

```python
from scipy.stats import zscore

X_train, X_test, y_train, y_test = train_test_split(X_SM, y_SM, test_size=0.3, random_state=11)
X_train_scaled= X_train.apply(zscore)
X_test_scaled= X_test.apply(zscore)
X_scaled=X.apply(zscore)
```

```python
#4.D.Check if the train and test data have similar statistical characteristics when compared to original data[2Marks]
from scipy.stats import ttest_ind
def test_stats(sample1,sample2):
    stats,p_value=ttest_ind(sample1,sample2,equal_var=False)
    if p_value.all()<0.05:
        print("Reject H0 under the significance level of 0.05 \n The two sample mean are significantly different ")
    else:
        print("Failed to reject the H0 \n The two sample mean are same")
```

```python
print("Comparing X and X_train_scaled")
test_stats(X_scaled,X_train_scaled)
print("")

print("Comparing X and X_test_scaled")
test_stats(X_scaled,X_test_scaled)
print("")

print("Comparing y and y_train")
test_stats(y,y_train)
print("")

print("Comparing y and y_test")
test_stats(y,y_test)
print("")
```
